```python
import librosa
def get_audio_duration(path):
    path = '/data4/data/personalize_audio2/' + path
    audio, sr = librosa.load(path)
    duration = len(audio)/sr
    return duration 
```

```python
from tqdm import tqdm

durations = []
for path in tqdm(personalize['local_path'], desc='Processing audio files'):
    duration = get_audio_duration(path)
    durations.append(duration)

personalize['duration'] = durations
```
